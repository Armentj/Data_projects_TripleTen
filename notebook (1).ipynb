### 2.2 Data Type Conversion

````python
# Make changes to data types if necessary
# Describe the columns where the data types have been changed and why.

# Convert 'year' to numeric (some entries may be non-numeric or missing)
games['year_of_release'] = pd.to_numeric(games['year_of_release'], errors='coerce').astype('Int64')

games['critic_score'] = pd.to_numeric(games['critic_score'], errors='coerce')

sales_columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
games[sales_columns] = games[sales_columns].astype(float)
````

````python
# Pay attention to the abbreviation TBD (to be determined). Specify how you intend to handle such cases.
# Replace all variations of TBD with NaN (case-insensitive)
games['user_score'] = games['user_score'].replace(['TBD', 'tbd'], np.nan)
# Replace 'TBD' with NaN
games['user_score'] = games['user_score'].replace('TBD', np.nan)

# Convert to float for analysis
games['user_score'] = games['user_score'].astype(float)
````

### 2.3 Handling Missing Values

````python
# Examine missing values
# Count missing values in each column
games.isna().sum()
````

````python
# Calculate percentage of missing values
missing_percent = games.isna().mean().round(4) * 100
missing_percent[missing_percent > 0]
````

````python
# Analyze patterns in missing values
# Create a DataFrame of boolean indicators for missing values
missing_df = games.isna()

# Correlation matrix of missingness
missing_corr = missing_df.corr()

# Visualize

plt.figure(figsize=(10, 8))
sns.heatmap(missing_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation of Missing Values')
plt.show()

````

````python
# Handle missing values based on analysis
# Your code here to handle missing values according to your strategy
# Drop rows where year_of_release is missing
games = games.dropna(subset=['year_of_release'])

# Calculate group means first, then merge back
group_means = games.groupby(['genre', 'platform'])['critic_score'].mean().reset_index()
group_means.columns = ['genre', 'platform', 'critic_score_mean']

# Merge with original data
games = games.merge(group_means, on=['genre', 'platform'], how='left')

# Fill missing values
games['critic_score'] = games['critic_score'].fillna(games['critic_score_mean'])

# Drop the helper column
games = games.drop('critic_score_mean', axis=1)
````

# Why do you think the values are missing? Give possible reasons.
# Explain why you filled in the missing values as you did or why you decided to leave them blank.
1. a lot of the missing values seem to be data entry errors or just incomplete records
2. Year_of_release missing values were dropped because year is essential for time-series analysis, trend detection, and filtering. Imputing a year could introduce misleading temporal patterns.
3. critic score was imputed with genre-platform mean. This Balances specificity and generalization genre and platform are strong predictors of review behavior. Avoids bias from global averages while preserving comparability. Keeps the dataset usable for modeling without discarding valuable rows.

## 3.2 Platform Sales Analysis Over Time

Now let's analyze how sales vary across platforms and years:

````python
# Calculate total sales by platform and year
# Group by platform and year, then sum total_sales
sales_by_platform_year = games.groupby(['platform', 'year_of_release'])['total_sales'].sum().reset_index()

# Sort for readability (optional)
sales_by_platform_year = sales_by_platform_year.sort_values(by=['year_of_release', 'total_sales'], ascending=[True, False])

# Display the result
sales_by_platform_year.head()
````

````python
# Create a heatmap of platform sales over time
# Step 1: Group and pivot the data
sales_pivot = games.groupby(['platform', 'year_of_release'])['total_sales'].sum().unstack(fill_value=0)

# Step 2: Set up the plot
plt.figure(figsize=(14, 8))
sns.heatmap(sales_pivot, cmap='YlGnBu', linewidths=0.5, linecolor='gray')

# Step 3: Customize labels
plt.title('Total Sales by Platform and Year', fontsize=16)
plt.xlabel('Year of Release')
plt.ylabel('Platform')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Step 4: Show the plot
plt.tight_layout()
plt.show()
````

````python
# Identify platforms with declining sales
# Group and sort sales data
sales_by_platform_year = games.groupby(['platform', 'year_of_release'])['total_sales'].sum().reset_index()
sales_by_platform_year = sales_by_platform_year.sort_values(by=['platform', 'year_of_release'])

# Calculate year-over-year change for each platform
sales_by_platform_year['sales_change'] = sales_by_platform_year.groupby('platform')['total_sales'].diff()

# Flag years with negative sales change
sales_by_platform_year['decline'] = sales_by_platform_year['sales_change'] < 0

# Count consecutive declines per platform
decline_counts = sales_by_platform_year.groupby('platform')['decline'].sum().sort_values(ascending=False)

# Display platforms with most years of decline
decline_counts.head()

#Line plot for top declining platforms
top_decliners = decline_counts.head(5).index
plt.figure(figsize=(12, 6))
for platform in top_decliners:
    data = sales_by_platform_year[sales_by_platform_year['platform'] == platform]
    plt.plot(data['year_of_release'], data['total_sales'], label=platform)

plt.title('Sales Trends for Declining Platforms')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend()
plt.tight_layout()
plt.show()
````

## 3.8 Genre Analysis

Finally, let's examine the distribution of games by genre:

````python
# Analyze genre performance
# Group by genre and calculate total sales and game count
genre_stats = (
    games_relevant.groupby('genre')
    .agg(game_count=('name', 'count'), total_sales=('total_sales', 'sum'))
    .reset_index()
)

# Calculate average sales per game
genre_stats['avg_sales_per_game'] = genre_stats['total_sales'] / genre_stats['game_count']
````

````python
# Sort genres by total sales
# Sort genres by total sales descending
genre_sorted = genre_stats.sort_values(by='total_sales', ascending=False)
````

````python
# Visualize genre distribution
plt.figure(figsize=(12, 6))
sns.barplot(data=genre_sorted, x='genre', y='total_sales', palette='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.title('Total Sales by Genre')
plt.ylabel('Total Sales (Millions)')
plt.xlabel('Genre')
plt.tight_layout()
plt.show()

````

````python
# Calculate market share for each genre
# Calculate market share as percentage of total sales
total_market_sales = genre_stats['total_sales'].sum()
genre_stats['market_share_pct'] = (genre_stats['total_sales'] / total_market_sales) * 100

# Sort by market share
genre_market_share = genre_stats.sort_values(by='market_share_pct', ascending=False)
genre_market_share[['genre', 'market_share_pct']].head()

# Calculate mean and median sales per genre from original data
genre_stats['mean_sales'] = games_relevant.groupby('genre')['total_sales'].mean().values
genre_stats['median_sales'] = games_relevant.groupby('genre')['total_sales'].median().values

genre_market_share = genre_stats.sort_values(by='market_share_pct', ascending=False)

# Display top genres with market share, mean, and median sales
genre_market_share[['genre', 'market_share_pct', 'mean_sales', 'median_sales']].head()
````

## 4.1 Regional Platform Analysis

Let's begin by examining platform performance across different regions:

````python
# Function to analyze platform performance by region
def analyze_platform_performance_by_region(df, region_column='Region', platform_column='Platform', sales_column='Sales'):
    """
    Analyzes platform performance by region.
    
    Parameters:
    - df: pandas DataFrame containing game sales data
    - region_column: column name for region data (e.g., 'Region')
    - platform_column: column name for platform data (e.g., 'Platform')
    - sales_column: column name for sales or revenue data (e.g., 'Sales')
    
    Returns:
    - summary_df: DataFrame showing total and average sales per platform per region
    """
    
    # Group by region and platform, then aggregate sales
    grouped = df.groupby([region_column, platform_column])[sales_column].agg(['sum', 'mean', 'count']).reset_index()
    
    # Rename columns for clarity
    grouped.columns = [region_column, platform_column, 'Total_Sales', 'Average_Sales', 'Game_Count']
    
    # Sort by region and total sales descending
    summary_df = grouped.sort_values(by=[region_column, 'Total_Sales'], ascending=[True, False])
    
    return summary_df
````

## 4.3 ESRB Rating Impact Analysis

Finally, let's examine how ESRB ratings affect sales in each region:

````python
# Function to analyze ESRB rating impact
def analyze_esrb_impact_by_region(df, rating_column='rating'):
    """
    Analyzes ESRB rating impact on sales across NA, EU, and JP.
    
    Returns a combined DataFrame with total and average sales per ESRB rating per region.
    """
    regions = {
        'NA': 'na_sales',
        'EU': 'eu_sales',
        'JP': 'jp_sales'
    }
    
    results = []
    
    for region_name, sales_column in regions.items():
        if rating_column not in df.columns or sales_column not in df.columns:
            continue  # Skip if required columns are missing
        
        grouped = (
            df.groupby(rating_column)[sales_column]
            .agg(['sum', 'mean', 'count'])
            .reset_index()
        )
        grouped.columns = [rating_column, 'Total_Sales', 'Average_Sales', 'Game_Count']
        grouped['Region'] = region_name
        results.append(grouped)
    
    # Combine results
    summary_df = pd.concat(results, ignore_index=True)
    return summary_df.sort_values(by=['Region', 'Total_Sales'], ascending=[True, False])

summary = analyze_esrb_impact_by_region(games_relevant)
print(summary)
````

````python
from scipy.stats import ttest_ind

# Null and Alternative Hypotheses (Test 1)
print("Test 1: Xbox One vs PC User Ratings")
print("Null Hypothesis (H‚ÇÄ): The average user ratings for Xbox One and PC are the same.")
print("Alternative Hypothesis (H‚ÇÅ): The average user ratings for Xbox One and PC are different.\n")

# Filter data
xbox_ratings = games_relevant[games_relevant['platform'] == 'XOne']['user_score'].dropna()
pc_ratings = games_relevant[games_relevant['platform'] == 'PC']['user_score'].dropna()

# Perform t-test
t_stat1, p_val1 = ttest_ind(xbox_ratings, pc_ratings, equal_var=False)

# Set alpha
alpha = 0.05

# Print results
print(f"t-statistic: {t_stat1:.4f}")
print(f"p-value: {p_val1:.4f}")
if p_val1 < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Üí Significant difference in average user ratings between Xbox One and PC.\n")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Üí No significant difference in average user ratings between Xbox One and PC.\n")

# Null and Alternative Hypotheses (Test 2)
print("Test 2: Action vs Sports Genre User Ratings")
print("Null Hypothesis (H‚ÇÄ): The average user ratings for Action and Sports games are the same.")
print("Alternative Hypothesis (H‚ÇÅ): The average user ratings for Action and Sports games are different.\n")

# Filter data
action_ratings = games_relevant[games_relevant['genre'] == 'Action']['user_score'].dropna()
sports_ratings = games_relevant[games_relevant['genre'] == 'Sports']['user_score'].dropna()

# Perform t-test
t_stat2, p_val2 = ttest_ind(action_ratings, sports_ratings, equal_var=False)

# Print results
print(f"t-statistic: {t_stat2:.4f}")
print(f"p-value: {p_val2:.4f}")
if p_val2 < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Üí Significant difference in average user ratings between Action and Sports genres.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Üí No significant difference in average user ratings between Action and Sports genres.")
````
